## Problem 5: The Longest train starting from Bangalore

Find the longest train starting from Bangalore.

````python
def top_distance(origin):
    sorted_values = df_trains[df_trains.from_station_code == origin].sort_values('duration_total',ascending=False).head(1)[["number","name","type","zone","from_station_code","from_station_name","to_station_code","to_station_name","departure","arrival","duration_h","duration_m","duration_total","distance"]]
    return sorted_values
````

````python
top_distance("SBC")
````

### Problem 6: Stations connected to Bangalore

Find all the stations connected to Bangalore by train. We are considering the stations that have a train starting from Bangalore and terminating in that station. 

````python
def connection(source):
    sorted_values = df_trains[(df_trains.from_station_code == source) | (df_trains.to_station_code == source)].sort_values('duration_total',ascending=False)[["number","name","type","zone","from_station_code","from_station_name","to_station_code","to_station_name","departure","arrival","duration_h","duration_m","duration_total","distance"]]
    return sorted_values
````

````python
connection("SBC")
````
