## Writing Files

Let's see how to write files


```python
f = open("a.txt", "w")
f.write("one\n")
f.write("two\n")
f.write("three\n")
f.close()
```

Remember to close and use w is somehow very important

```python
open("a.txt").read()
```

We could use "with" statement ot automatically close the file

```python
with open ("a.txt", "w") as f:
    f.write("one\n")
    f.write("two\n")
    f.write("three\n")
```

We can also open a file in append movde by specifying "a" for mode

```python
# append more contents to the file a.txt
with open ("a.txt","a") as f:
    f.write("four\n")
```

### Figlet API

```python
!curl 'https://figlet.apps.pipal.in/api/figlet?text=Python'
```

```python
url = 'https://figlet.apps.pipal.in/api/figlet?text=Python'
```

```python
output = requests.get(url).text
print(output)
```

The part '?text=Python' is a query string. This is how we pass inputs to a web request.

The 'requests' library calls it 'params' and allows specifying them separately

```python
url = "https://figlet.apps.pipal.in/api/figlet"
params = {"text": "FUN"}

output = requests.get(url, params=params).text
print(output)
```

```python
def figlet(text):
    url = "https://figlet.apps.pipal.in/api/figlet"
    params = {"text": text}

    return requests.get(url, params=params).text
```

```python
print(figlet("Python is awesome"))
```

### JSON APIs

JSON is a popular serialization format for data.

Python has a 'json' module for handling it, but the requests library takes care of it behind the scenes

```python
import json
```

```python
# value = [1,2,3,4]
value = {"name": "Alice", "phone": 1234, "x": [1,2,3]}
jsontext = json.dumps(value)
jsontext
```

```python
print(jsontext)
```

```python
# python dictionary
d = json.loads(jsontext)
print(d['name'])
```

### Example: Frankfurter Currency Conversion

```python
def convert_currency(symbol, base="USD"):
    url = "https://api.frankfurter.app/latest"
    params = {
    "base": "USD",
    "date": "2024-10-18", "symbols": "INR"
     }
    data = requests.get(url, params=params).json()
    return data['rates'][symbol]
```

```python
convert_currency("INR")
```
